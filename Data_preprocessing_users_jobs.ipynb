# Jobs and Users statistics

```python
users = slurm1[['UID','Submit']]
users['year'] = pd.DatetimeIndex(users['Submit']).year
users['month'] = pd.DatetimeIndex(users['Submit']).month
users['year'] = users['year'].apply(str)
users['month'] = users['month'].apply(str)
users['year_month'] = users['year'] + '-' + users['month']
number_of_jobs = Counter(users['year_month'])
dict_month_uid = {}
for index,row in users.iterrows():
    
    str_month = str(row['year_month'])
    str_uid_month = str(row['UID'])
    set_temp_month=dict_month_uid.get(str_month, set())
    set_temp_month.add(str_uid_month)
    dict_month_uid[str_month] = set_temp_month
#getting the number of users per month
dict_month_count_uid = {}
for key,value in dict_month_uid.items():
    dict_month_count_uid[key] = len(value)
```

```python
def function1(years,choices):
    check = 0
    coun = 0
    for c in choices:
        n = len(years)
        lists = np.empty((n, 0)).tolist()
        listst = []
        listave = []
        for i in years:
            for j in lists:
                if not j:
                    temp = {k:v for (k,v) in c.items() if i in k}
                    temp1 = list(temp.values())
                    length= int(len(temp1))
                    mean = sum(temp1) / length 
                    ans = sum((i - mean) ** 2 for i in temp1) / length
                    std = math.sqrt(ans)
                    listave.append(std)
                    temp1 = [y - x for x, y in zip(temp1[:-1], temp1[1:])]
                    listst.append(temp1)
                    j.append(temp)
                    break
        listst = [x for x in listst if x != []]
        for l in lists:
            l = dict(ChainMap(*l))
            count = 0
            fig, (ax, ax_table) = plt.subplots(nrows=2)
            ax_table.axis('off')
            if check <= 3:
                ax.set_title("Number of jobs per month" ,fontsize=20)
                check = check + 1
            elif check >= 4:
                ax.set_title("Number of users per month", fontsize=20)
                check = check + 1                                                                 
            width = .5 
            ax.bar(l.keys(), l.values(), width, color='g')
            ax.set_xticklabels(l.keys(), fontsize=16)
            ax.set_yticklabels(l.values(), fontsize=16)
            fig.set_figheight(20)
            fig.set_figwidth(18)
            for p in ax.patches:
                ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005),size = 18)
            if len(listst) != 0:
                for index,l1 in enumerate(listst):
                    for index,li in enumerate(listave):
                        if count == 1:
                            break
                        print("Standard Deviation {}".format(li))
                        ax_table = plt.table(cellText=[l1],rowLabels=['Difference'],loc= 1)
                        ax_table.set_fontsize(14)
                        ax_table.scale(0.9, 1.0)
                        props = dict(boxstyle='round', facecolor='wheat', alpha=0.2)
                        textstr = '\n'.join((
                        r'$SD= %.2f$' % (li),))
                        ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=16,
                        verticalalignment='top', bbox=props)
                        plt.subplots_adjust(left=0, bottom=0)
                        count = count + 1
                        listst.pop(0)
                        listave.pop(0)
            plt.show()
            coun = coun + 1
            if not os.path.exists('no_of_jobs_and_users_month'):
                os.makedirs('no_of_jobs_and_users_month')
            fig.savefig('no_of_jobs_and_users_month/{}.png'.format(str(coun)),bbox_inches='tight', dpi=100)
        
```

```python
#Printing the jobstep patterns to match - 6 main types
import re
scc = []
chars = ''
alphabets = []
for ele in JobID:
    pat = re.sub(r'[0-9 a-z]*', '', str(ele))
    if pat not in scc:
        scc.append(pat)
    words = ''.join(re.findall("[a-zA-Z]+", str(ele)))
    if words not in alphabets:
        alphabets.append(words)
print(alphabets)
print(scc)
```

```python
#Seperating all the 6 job step types into different lists
count_pat1 = 0
count_pat2 = 0
count_pat3 = 0
count_pat4 = 0
count_pat5 = 0
count_pat6 = 0
count_pat7 = 0
import re
type1 = [] 
type2 = []
type3 = []
type4 = []
type5 = []
type6 = []
type7 = []
dict_type1 = {}
dict_type2 = {}
for index, elem in enumerate(JobID):
    nextele = JobID[(index +1) % len(JobID)]
    pattern5 = re.compile(r'^[0-9]*$')
    pattern6 = re.compile(r'^[0-9]*\_[0-9]*$')
    pattern1 = re.compile(r'^[0-9]*\.[a-z 0-9]*$')
    pattern2 = re.compile(r'^[0-9]*\_[0-9]*\.[a-z 0-9]*$')
    pattern3 = re.compile(r'^[0-9]*\_[0-9]*\[[0-9]*\-[0-9]*\]$')
    pattern4 = re.compile(r'^[0-9]*\_[0-9]*\[[0-9]*\]$')
    sep = '.'
    if pattern1.match(str(elem)):
        count_pat1 = count_pat1 + 1
        trimmedelmt1 = str(elem).split(sep, 1)[0]
        if trimmedelmt1 not in dict_type1:
            type1.append(trimmedelmt1)
            dict_type1[trimmedelmt1] = ''
        type1.append(elem)
        dict_type1[elem] = ''
    elif pattern2.match(str(elem)):
        count_pat2 = count_pat2 + 1
        trimmedelmt2 = str(elem).split(sep, 1)[0]
        if trimmedelmt2 not in dict_type2:
            type2.append(trimmedelmt2)
            dict_type2[trimmedelmt2] = ''
        type2.append(elem)
        dict_type2[elem] = ''
    elif pattern3.match(str(elem)):
        count_pat3 = count_pat3 + 1
        if elem not in type3:
            type3.append(elem)
    elif pattern4.match(str(elem)):
        count_pat4 = count_pat4 + 1
        if elem not in type4:
            type4.append(elem)
    elif pattern5.match(str(elem)):
        count_pat6 = count_pat6 + 1
        if '.' not in nextele:
            type6.append(elem)
    elif pattern6.match(str(elem)):
        if '.' not in nextele:
            type7.append(elem)
        count_pat7 = count_pat7 + 1
        
    else:
        type5.append(elem)
        count_pat5 = count_pat5 + 1
```
